````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Load dataset
iris = load_iris()

X = iris.data      # Features
y = iris.target    # Labels

print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


# Create KNN classifier (k=3)
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X_train, y_train)


y_pred = knn.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(f"K={k}, Accuracy={accuracy_score(y_test, y_pred):.3f}")
````

````python
Here‚Äôs a complete example of **Linear Regression using the Iris dataset** in Python.

> ‚ö†Ô∏è Note: Iris is mainly for classification, but we can use **Linear Regression** to predict a continuous variable like **sepal length**.

---

# üìà Linear Regression on Iris Dataset

We‚Äôll predict **sepal length** using the other three features.

---

## 1Ô∏è‚É£ Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

---

## 2Ô∏è‚É£ Load Dataset

```python
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df.head()
```

---

## 3Ô∏è‚É£ Define Features and Target

We will predict:

üëâ **Target (y): sepal length**
üëâ **Features (X): other 3 features**

```python
X = df[['sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y = df['sepal length (cm)']
```

---

## 4Ô∏è‚É£ Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

---

## 5Ô∏è‚É£ Train Linear Regression Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

---

## 6Ô∏è‚É£ Make Predictions

```python
y_pred = model.predict(X_test)
```

---

## 7Ô∏è‚É£ Evaluate Model

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Sepal Length")
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
````

---

## 2Ô∏è‚É£ Load Dataset

````python
iris = load_iris()

# Convert to DataFrame
df_iris_linear = pd.DataFrame(iris.data, columns=iris.feature_names)

display(df_iris_linear.head())
````

We will predict:

---
